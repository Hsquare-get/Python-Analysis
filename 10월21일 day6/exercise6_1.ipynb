```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

df = pd.read_csv('../data/cctv_seoul.csv')
# display(df.info())
# display(df.head())

%config InlineBackend.figure_format='retina' #   

font_path = "../data/THEdog.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

Y = df['']
WIDTH = df['CCTV']

plt.figure(figsize=(10,10))
plt.margins(y=0.01)
plt.title(" (1)   CCTV  ", fontdict={'fontsize':20})
plt.xlabel('   CCTV ', fontdict={'fontsize':15})
plt.ylabel('', fontdict={'fontsize':15})
plt.grid()

mycolors = sns.color_palette('Spectral', len(Y))
plt.barh(y=Y, width=WIDTH, color=mycolors)
plt.savefig('./hw1.png') # plt.savefig()  plt.show()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

df = pd.read_csv('../data/cctv_seoul.csv')
# display(df.info())
# display(df.head())
df.sort_values(by='CCTV', ascending=False, inplace=True)

%config InlineBackend.figure_format='retina' #   

font_path = "../data/THEdog.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

X = df['']
Y = df['CCTV']

plt.figure(figsize=(20,10))
plt.margins(x=0.01, y=0.05)
plt.title(" (2)   CCTV  ", fontdict={'fontsize':20})
plt.xlabel('', fontdict={'fontsize':15})
plt.xticks(rotation=20)
plt.ylabel('   CCTV ', fontdict={'fontsize':15})
plt.grid()

mycolors = sns.color_palette('PRGn', len(X))
plt.bar(x=X, height=Y, color=mycolors)
plt.savefig('./hw2.png') # plt.savefig()  plt.show()
plt.show()
```
