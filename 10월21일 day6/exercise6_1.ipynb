```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

df = pd.read_csv('../data/cctv_seoul.csv')
# display(df.info())
# display(df.head())

%config InlineBackend.figure_format='retina' # 화면 해상도 향상

font_path = "../data/THEdog.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

Y = df['구별']
WIDTH = df['CCTV수']

plt.figure(figsize=(10,10))
plt.margins(y=0.01)
plt.title("시각화 과제(1) 각 구의 CCTV 설치 현황", fontdict={'fontsize':20})
plt.xlabel('각 구에 설치된 CCTV 댓수', fontdict={'fontsize':15})
plt.ylabel('구이름', fontdict={'fontsize':15})
plt.grid()

mycolors = sns.color_palette('Spectral', len(Y))
plt.barh(y=Y, width=WIDTH, color=mycolors)
plt.savefig('./hw1.png') # plt.savefig()로 저장하고 plt.show()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

df = pd.read_csv('../data/cctv_seoul.csv')
# display(df.info())
# display(df.head())
df.sort_values(by='CCTV수', ascending=False, inplace=True)

%config InlineBackend.figure_format='retina' # 화면 해상도 향상

font_path = "../data/THEdog.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

X = df['구별']
Y = df['CCTV수']

plt.figure(figsize=(20,10))
plt.margins(x=0.01, y=0.05)
plt.title("시각화 과제(2) 각 구의 CCTV 설치 현황", fontdict={'fontsize':20})
plt.xlabel('구이름', fontdict={'fontsize':15})
plt.xticks(rotation=20)
plt.ylabel('각 구에 설치된 CCTV 댓수', fontdict={'fontsize':15})
plt.grid()

mycolors = sns.color_palette('PRGn', len(X))
plt.bar(x=X, height=Y, color=mycolors)
plt.savefig('./hw2.png') # plt.savefig()로 저장하고 plt.show()
plt.show()
```
