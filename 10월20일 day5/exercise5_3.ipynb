```python
import pandas as pd

file_path = '../data/score.csv'
df = pd.read_csv(file_path)

print("Columns :", df.columns)
print("Index :", df.index)

df['total'] = df['kor'] + df['eng'] + df['mat']
df['avg'] = df['total'] / 3
pd.options.display.float_format = '{:.2f}'.format # 소수 둘째자리까지 표기
# pd.reset_option('display.float_format') # 설정 초기화
display(df)
```

```python
import pandas as pd

file_path = '../data/score_noheader.csv'
df = pd.read_csv(file_path, header=None)
display(df)

print("Columns :", df.columns)
print("Index :", df.index)

df.rename(columns={0:'name', 1:'kor', 2:'eng', 3:'mat'}, inplace=True)
print("Columns :", df.columns)
display(df)

df['total'] = df.kor + df.eng + df.mat
df['avg'] = df.total / 3
display(df)
```

```python
import pandas as pd

file_path = '../data/score.xlsx'
df = pd.read_excel(file_path)
display(df)

df['total'] = df.kor + df.eng + df.mat
df['avg'] = df['total'] / 3
df.sort_values(by='avg', ascending=False, inplace=True)
display(df)

df.to_excel('./score_result1.xlsx', index=True)
df.to_excel('./score_result2.xlsx', index=False)
```
